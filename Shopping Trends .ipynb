```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('shopping_trends.csv')
```

```python
df.head()
```

```python
# Plot a histogram of purchase amount
df['Purchase Amount (USD)'].plot(kind='hist' , edgecolor='black')
plt.title('Distribution of Purchas Amount')
plt.xlabel('Purchase Amount (USD)')
plt.show()
```

```python
#purchase amount by gender
sns.countplot(data = df, x = 'Gender')
plt.title('Count of F/M')
df['Gender'].value_counts() 
```

```python
# Bar chart of the most popular item categories
top_categories = df['Category'].value_counts()
top_categories.plot(kind='bar', color='skyblue' )
plt.title('Top 10 Popular Item Categories')
plt.xlabel('Item Category')
plt.ylabel('Count')
plt.show()
```

```python
# Line plot of average purchase amount over seasons
seasonal_purchase = df.groupby('Season')['Purchase Amount (USD)'].mean()
seasonal_purchase.plot(kind='line' )
plt.title('Average Purchase Amount Over Seasons')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```

```python
# Bar plot of  purchase amount over seasons
sns.barplot(x='Season', y='Purchase Amount (USD)', data=df)
```

```python
# Bar chart of the impact of discounts on purchase amount
promo_code_counts = df['Promo Code Used'].value_counts()
plt.pie(promo_code_counts, labels=promo_code_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Promo Code Usage')


```

```python
# Bar chart of average purchase amount by location

location_purchase = df.groupby('Location')['Purchase Amount (USD)'].mean()
location_purchase.plot(kind='bar'  ,figsize=(12, 6))
plt.title('Average Purchase Amount by Location')
plt.xlabel('Location')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```

```python
# Box plot of purchase amount by subscription status
sns.boxplot(x='Subscription Status', y='Purchase Amount (USD)', data=df  )
plt.title('Purchase Amount by Subscription Status')
plt.show()
```

```python
# Scatter plot of review ratings against purchase amount

sns.scatterplot(x='Review Rating', y='Purchase Amount (USD)', data=df)
plt.title('Review Ratings vs. Purchase Amount')
plt.xlabel('Review Rating')
plt.ylabel('Purchase Amount (USD)')
plt.show()

# Bar chart of average review ratings by category

category_ratings = df.groupby('Category')['Review Rating'].mean()

category_ratings.plot(kind='bar')
plt.title('Average Review Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Average Review Rating')
plt.show()
```

```python
# Scatter plot of previous purchases against current purchase amount

sns.scatterplot(x='Previous Purchases', y='Purchase Amount (USD)', data=df)
plt.title('Previous Purchases vs. Purchase Amount')
plt.xlabel('Previous Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.show()

# Box plot of purchase amount by frequency of purchases

sns.boxplot(x='Frequency of Purchases', y='Purchase Amount (USD)', data=df)
plt.title('Purchase Amount by Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.show()
```

```python
# Count plot of preferred payment methods
plt.figure(figsize=(10, 6))
sns.countplot(x='Payment Method', data=df)
plt.title('Preferred Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```
